```python
import pandas as pd
import numpy as np
import sklearn
from numpy.random import seed
import os
from sklearn import datasets, linear_model

from sklearn.model_selection import train_test_split
```

```python
wine_csv = os.path.join("resources", "winemag-data-130k-prices-only.csv")
wine_df = pd.read_csv(wine_csv).drop(columns="Unnamed: 0")

len(wine_df)
```

```python
wine_df = wine_df.dropna(subset=['variety'])
wine_df = wine_df.dropna(subset=['points'])
wine_df = wine_df.dropna(subset=['price'])
len(wine_df)
```

```python
varieties_array = wine_df['variety'].unique()
```

```python
regr = linear_model.LogisticRegression()

def run_reg_analysis(df):
    X = df['price'].values.reshape(-1,1)
    y = df['points'].values
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
    regr.fit(X_train, y_train)
    score = regr.score(X_test, y_test)
    if score > 0.8:
        print(f"{variety} is highly correlated, points to price (score is {score})")
    print(score)
```

```python
i = 1

for variety in varieties_array:
    selection = wine_df.loc[wine_df['variety'] == variety]
    if len(selection) > 10:
        run_reg_analysis(selection)
    print("Variety processed, on to the next one!")
```

```python
Riesling = wine_df.loc[wine_df['variety'] == "Riesling"]
print(Riesling)
```

```python
wine = wine_df.loc[wine_df['variety'] == "Riesling"]

run_reg_analysis(wine)
```
